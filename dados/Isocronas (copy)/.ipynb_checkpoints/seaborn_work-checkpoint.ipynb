```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('isocro.csv', header = 0, usecols = ['Mv', '(B-V)o', 'log(Age)'], nrows = 875)
data.tail()
```

```python
idades = data['log(Age)']
age = np.array([])
for element in idades:
    if element in age:
        continue
    elif element not in age:
        age = np.append(age,element)
```

```python
Mvarray = []
Bvarray = []
for element in age:
    a = data[data['log(Age)']==element]
    b = a['Mv'].tolist()
    c = a['(B-V)o'].tolist()
    Mvarray.append(b)
    Bvarray.append(c)
```
