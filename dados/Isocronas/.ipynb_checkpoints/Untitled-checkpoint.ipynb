```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('Regression_Iso.txt', header = 0)
data.tail()
```

```python
plt.rc('xtick', labelsize=13) 
plt.rc('ytick', labelsize=13) 


font = {'family' : 'sans-serif',
        'weight' : 'normal',
        'size'   : 13}

plt.rc('font', **font)


plt.figure(figsize=(10,8))
plt.errorbar( data['Age'], data['Slope'], yerr = data['Slope_Error'], fmt = 'o' , color = 'black', ecolor = 'red')
plt.ylabel('Slope')
plt.xlabel('Age');
```

```python
plt.rc('xtick', labelsize=13) 
plt.rc('ytick', labelsize=13) 


font = {'family' : 'sans-serif',
        'weight' : 'normal',
        'size'   : 13}

plt.rc('font', **font)


plt.figure(figsize=(10,8))
plt.errorbar(data['Age'], data['Intercept'],  yerr = data['Intercept_Error'], fmt = 'o' , color = 'black', ecolor = 'red')
plt.ylabel('Intercept')
plt.xlabel('Age');
```

```python
plt.rc('xtick', labelsize=13) 
plt.rc('ytick', labelsize=13) 


font = {'family' : 'sans-serif',
        'weight' : 'normal',
        'size'   : 13}

plt.rc('font', **font)



x = data['(B-V)TurnOff'] #para o scatter
plt.figure(figsize=(10,8))
plt.gca().invert_yaxis()
#plt.errorbar( data['Age'], data['(B-V)TurnOff'], fmt = 'o' , color = 'black', ecolor = 'red')
plt.scatter( data['Age'], data['(B-V)TurnOff'],  c = x, cmap = 'bwr') #para reverter o colormap, bwr_r, _r means reversed!
plt.ylabel('(B-V) TurnOff Point')
plt.xlabel('Age')
plt.colorbar();
```

```python
plt.rc('xtick', labelsize=13) 
plt.rc('ytick', labelsize=13) 


font = {'family' : 'sans-serif',
        'weight' : 'normal',
        'size'   : 13}

plt.rc('font', **font)

x = data['MvTurnOff']
plt.figure(figsize=(10,8))
plt.gca().invert_yaxis()
#plt.errorbar(data['Age'], data['MvTurnOff'], fmt = 'o' , color = 'black', ecolor = 'red')
plt.scatter( data['Age'], data['MvTurnOff'],  c = x, cmap = 'bwr')
plt.ylabel('Mv TurnOff Point')
plt.xlabel('Age');
plt.colorbar();
```
