```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from astroquery.gaia import Gaia
import astropy
```

```python
tables = Gaia.load_tables(only_names=True)
```

```python
meta = Gaia.load_table('gaiadr3.gaia_source')
meta
```

```python
df = pd.read_csv('final_5d.csv')

ids = df.source_id

query_ids = ""
for i in range(len(ids)):
  if i == len(ids)-1:
    query_ids += 'source_id = '+ str(ids[i])
  else:
   query_ids += 'source_id = '+ str(ids[i]) + ' OR '
```

```python
query = """SELECT 
source_id, ra_error, dec_error, ra_dec_corr, ra_parallax_corr, ra_pmra_corr, ra_pmdec_corr, dec_parallax_corr, dec_pmra_corr, dec_pmdec_corr,
parallax_pmra_corr, parallax_pmdec_corr, pmra_pmdec_corr
FROM gaiadr3.gaia_source
WHERE """   + query_ids
```

```python
job = Gaia.launch_job_async(query)
job
```

```python
data = job.get_results()
```

```python
extra = data.to_pandas()
```

```python
if True in extra.isnull():
    print('uepa')
```

```python
dados = pd.merge(df,extra)
dados
```

```python
RA = np.arange(min(dados.ra),max(dados.ra), 3*np.mean(dados.ra_error))
DEC = np.arange(min(dados.dec),max(dados.dec),3*np.mean(dados.dec_error))
PMRA = np.arange(min(dados.pmra),max(dados.pmra) + 10*np.mean(dados.pmra_error) ,5*np.mean(dados.pmra_error))
PMDEC = np.arange(min(dados.pmdec),max(dados.pmdec) + 10*np.mean(dados.pmdec_error), 5*np.mean(dados.pmdec_error))
PLX = np.arange(min(dados.parallax),max(dados.parallax) + np.mean(dados.parallax_error),np.mean(dados.parallax_error))

len(RA),len(DEC),len(PMRA),len(PMDEC), len(PLX)
```

```python
from numpy.linalg import inv, det
probs = np.zeros(len(dados))

for i in range(len(dados)):
        star = dados.loc[i]
        m = np.array([[star.pmra_error**2,star.pmra_error*star.pmdec_error*star.pmra_pmdec_corr,star.pmra_error*star.parallax_error*star.parallax_pmra_corr],
        [star.pmra_error*star.pmdec_error*star.pmra_pmdec_corr,star.pmdec_error**2,star.pmdec_error*star.parallax_error*star.parallax_pmdec_corr], 
        [star.pmra_error*star.parallax_error*star.parallax_pmra_corr,star.pmdec_error*star.parallax_error*star.parallax_pmdec_corr,star.parallax_error**2]])
        matriz = m
        pmras = []
        for j in range(len(PMRA)):
            if star.pmra <= PMRA[j]:
                pmras.append(PMRA[j-1])
                pmras.append(PMRA[j])
                break
            else:
                continue
        pmdecs = []
        for j in range(len(PMDEC)):
            if star.pmdec <= PMDEC[j]:
                pmdecs.append(PMDEC[j-1])
                pmdecs.append(PMDEC[j])
                break
            else:
                continue
        plxs = []
        for j in range(len(PLX)):
            if star.parallax <= PLX[j]:
                plxs.append(PLX[j-1])
                plxs.append(PLX[j])
                break
            else:
                continue
        subset = dados.loc[#(np.min(ras) <= dados.ra) & (dados.ra<=np.max(ras)) &
                        #(np.min(decs) <= dados.dec) & (dados.dec<=np.max(decs)) &
                        (np.min(pmras) <= dados.pmra) & (dados.pmra<=np.max(pmras)) &
                        (np.min(pmdecs) <= dados.pmdec) & (dados.pmdec<=np.max(pmdecs)) &
                        (np.min(plxs) <= dados.parallax) & (dados.parallax<=np.max(plxs))]
        X = np.array([[star.pmra*np.cos(np.deg2rad(star.dec))],[star.pmdec],[star.parallax]])
        MEDIAS = np.array([[subset.pmra.mean()],[subset.pmdec.mean()],[subset.parallax.mean()]])
        matrizinv = inv(matriz)
        determinante = det(matriz)
        l = np.exp(-1/2*np.matmul((X-MEDIAS).T, np.matmul(matrizinv, (X-MEDIAS))))/(np.sqrt(((2*np.pi)**3)*determinante))
        probs[i] = l
np.count_nonzero(probs>0)
```
