```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('MEMBERS-CANTAT.tsv',sep = ';', comment = '#')
dados['BP-RP'] =  pd.to_numeric(dados['BP-RP'], errors='coerce')
dados['Gmag'] =  pd.to_numeric(dados['Gmag'], errors='coerce')
len(dados)
```

```python
df = dados[dados['Cluster']=='Berkeley_29']
print(len(df))
plt.scatter(df ['BP-RP'], df['Gmag'])
plt.gca().invert_yaxis()
```
