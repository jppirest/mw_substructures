```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.odr import * 
```

```python
data = pd.read_csv('isocro.csv', header = 0)
data.tail()
```

```python
idades = data['log(Age)']
age = np.array([])
for element in idades:
    if element in age:
        continue
    elif element not in age:
        age = np.append(age,element)

list_dataframes = []
for element in age:
    list_dataframes.append(data[data['log(Age)']==element])
```

```python
nova = data[data['log(Age)']==6.6]
Mv = nova['Mv']
x = nova['(B-V)o'] 
y = -Mv + (x-Mv)
x = nova['(B-V)o'] 
plt.scatter(x,y, color = 'none', edgecolor = 'blue')
#plt.plot(x,y, '--')
```

```python
x=[]
y=[]
for i in list_dataframes:
    y.append(i['Mv'])
    x.append(i['(B-V)o'])
    
```

```python
plt.figure(figsize=(7,5.5))
for a,b in zip(x,y):
    plt.plot(a,-b +(a-b))
    plt.xlabel(r"$( B-V)_o$")
    plt.ylabel(r"$V$")
    plt.tight_layout()
```
